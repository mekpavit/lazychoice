```python
restaurants = []
for page_num in range(1, 11):
    driver.get('https://www.wongnai.com/businesses?nearBusinessId=329382&spatialInfo.radius=0.62&domain=1&rerank=true&page.number={}'.format(page_num))
    html_content = driver.page_source
    soup = BeautifulSoup(html_content, 'html.parser')
    contents = soup.find_all(class_=re.compile("content"))[1].contents[0].find_all(class_='row')[0].find_all(class_='col-md-6')[0].contents[0]
    for ele in contents.find_all(name=['a']):
        restaurant_url = ele.get('href')
        try:
            restaurant_name = ele.find_all(name=['h2', 'h5'])[0].string
            restaurants.append({
                'name': restaurant_name,
                'url': 'https://www.wongnai.com' + restaurant_url,
                'location': 'BTS Siam'
            })
        except:
            pass
for page_num in range(1, 11):
    driver.get('https://www.wongnai.com/places/bts-chongnon-si?domain=1&regions=11164&page.number={}'.format(page_num))
    html_content = driver.page_source
    soup = BeautifulSoup(html_content, 'html.parser')
    contents = soup.find_all(class_=re.compile("content"))[1].contents[0].find_all(class_='row')[0].find_all(class_='col-md-6')[0].contents[0]
    for ele in contents.find_all(name=['a']):
        restaurant_url = ele.get('href')
        try:
            restaurant_name = ele.find_all(name=['h2', 'h5'])[0].string
            restaurants.append({
                'name': restaurant_name,
                'url': 'https://www.wongnai.com' + restaurant_url,
                'location': 'BTS Chong nonsi'
            })
        except:
            pass
```

```python
import json
with open('./restaurants.json', 'w') as f:
    f.write(json.dumps(restaurants))
```
